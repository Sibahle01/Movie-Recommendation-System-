```python
movies=movies[['id', 'title', 'overview', 'genre']]
```

```python
movies['tags'] = movies['overview']+movies['genre']
```

Now We remove overview and genre columns because we now have tags calumn which is the mixture of both

```python
new_data = movies.drop(columns=['overview', 'genre'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data.info()
```

Calculate The Movies with the Highest Cosine Similarities

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

Create a function that only Recomend us Movie Based on the Title of the movie not the ID Just the Title

```python
def recommend(movies):
    index = new_data[new_data['title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    for i in distance[0:5]:
        print(new_data.iloc[i[0]].title)
```

```python
recommend("Iron Man")
```
